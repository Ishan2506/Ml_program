```python
#SECTION-B

#Q1
df1=pd.read_csv('/content/fracture.csv')
df1.head(15)
```

```python
#Q2
df1['bmi'] = df1['weight_kg'] / (df1['height_cm'] / 100) ** 2
df1
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label=LabelEncoder()
df1['sex']=label.fit_transform(df1['sex'])
df1
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df1[['age', 'sex', 'bmi', 'bmd']].values
y = df1['fracture'].values

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
x_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
train_predictions = model.predict(x_train)
test_predictions = model.predict(x_test)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(train_predictions)
plt.title('Train Predictions')
plt.xlabel('Fracture')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(test_predictions)
plt.title('Test Predictions')
plt.xlabel('Fracture')
plt.ylabel('Count')
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```
